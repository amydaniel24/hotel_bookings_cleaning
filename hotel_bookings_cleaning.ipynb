```python
import pandas as pd

# Load, read add get the first few rows of the dataset
file_path = "data/hotel_bookings.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
#shape and columns
print("Rows, columns:", df.shape)
print("\nColumn names:")
print(df.columns.tolist())
```

```python
# info and describe
df.info()
df.describe()
```

```python
# value checks
print(df['hotel'].value_counts())
print()
print(df['is_canceled'].value_counts())
print()
print(df['market_segment'].value_counts())
```

```python
# missing values
df.isna().sum().sort_values(ascending=False)
```

```python
# remove duplicates
df.duplicated().sum()
df = df.drop_duplicates().reset_index(drop=True)
```

```python
# Remove bookings with no guests
df['total_guests'] = df['adults'] + df['children'].fillna(0) + df['babies']
df[df['total_guests'] == 0].shape
df = df[df['total_guests'] > 0].reset_index(drop=True)
```

```python
# Fill missing values in children with 0
df['children'] = df['children'].fillna(0)
```
